```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips,y="tip",x="total_bill",hue="day",style="time",palette="pastel",kind="line")#hue is used to create difference based on a column via colors
```

```python
sns.catplot(data=tips,x="day",y="total_bill",kind="violin",hue="smoker")
```

```python
sns.catplot(data=tips,x="total_bill",y="day")
```
