```python
#plot the score of two students in five different subject subject as x axis and marks have y axis
s1=[56,78,92,95,78]
s2=[77,89,91,23,45]
sub=["maths","science","english","telugu","hindi"]
plt.plot(sub,s1,label="s1",color="c");
plt.plot(sub,s2,label="s2",color="b");
plt.title("marks of subject")
plt.legend()
```

```python
import numpy as np
```

```python
a=np.array([230,560,780,127,128])
b=np.array([200,160,270,127,400])
print("revenues of company A",a)
print("revenues of company B",b)
profa=np.diff(a)
print("profit of A",profa)
profb=np.diff(b)
print("profit of B",profb)
year=["2019-20","2020-21","2021-2022","2022-2023"]
plt.subplot(2,1,1)
plt.bar(year,profa,label="profit of A",color="black")
plt.legend()
plt.subplot(2,1,2)
plt.bar(year,profb,label="profit of B",color="g")
plt.legend()
```

```python
a=np.array([26,60,5,10])
labe=["AIML","PYTHON","PANDAS","NUMPY"]
plt.pie(a,labels=labe)
plt.show
```

```python
a=np.array([26,60,5,10])
labe=["AIML","PYTHON","PANDAS","NUMPY"]
explo=[0.2,0,0,0]
plt.pie(a,labels=labe,explode=explo,startangle=180,shadow=True)
plt.legend()
plt.show
```
