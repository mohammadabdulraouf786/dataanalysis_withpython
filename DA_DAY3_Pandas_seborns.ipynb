hue= different category
style

```python
tips=sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="day")
#hue - creating difference based on a column via colors
```

```python
sns.relplot(data=tips,x="total_bill",y="tip",hue="sex",style="time")
```

```python
sns.catplot(data=tips,x="total_bill",y="day")
```

```python
sns.catplot(data=tips,y="total_bill",x="time",kind="violin",hue="smoker")
```

```python
dowjones=sns.load_dataset("dowjones")
dowjones.head()
```
